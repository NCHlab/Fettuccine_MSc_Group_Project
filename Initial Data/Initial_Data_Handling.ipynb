```python
#Get HERV superfamilies and families

#Read the file of nested ERVs from RepeatMasker using Pandas
import pandas as pd
import re
ervs = pd.read_csv("all_ERVS.csv", header = 1)
#Remove superfamilies for which the RepeatMasker is unsure
ervs_no_qerv1 = ervs[ervs.repFamily != "ERVL?"]
ervs_no_unsure = ervs_no_qerv1[ervs_no_qerv1.repFamily != "ERV1?"]
#Remove -int
#ervs_no_unsure['name'] = ervs_no_unsure['name'].apply(lambda x: re.sub('-int','',x))

#Function that finds and counts families for a given superfamily
def fams_to_sfams(sfamily):
    erv_names = ervs_no_unsure["repName"][ervs.repFamily == str(sfamily)]
    counts = erv_names.value_counts()
    
    #Create a dataframe with Superfamily, Family, Counts format
    fam_counts_df = pd.DataFrame(data=counts, index=None, columns=None, dtype=None, copy=False)
    fam_counts_df.columns = ["Count"]
    fam_counts_df.reset_index(level=0, inplace=True)
    fam_counts_df.insert(0,"Superfamily", str(sfamily))
    fam_counts_df = fam_counts_df.rename(columns={'index': 'Family'}) 
    
    return fam_counts_df

#Create a dataframe families_data with Families and 
#how many time they were found in genome catergorized 
#by Superfamilies for the whole Data.
erv_superfamilies = ["ERV1","ERVL","ERVK", "ERVL-MaLR"]
families_data = pd.DataFrame(columns=["Superfamily","Family","Count"])

for sfam in erv_superfamilies:
    temp = fams_to_sfams(str(sfam))
    families_data = families_data.append(temp, ignore_index=False, verify_integrity=False)
    
families_data.to_csv(path_or_buf="L1_Families.csv", sep=' ')
```
